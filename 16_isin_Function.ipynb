#### Applying OR operator - Applying on Single Column

```python
cond_1 = df_new["AreaName"] == "Asia"
cond_2 = df_new["AreaName"] == "Africa"
df_new[cond_1 | cond_2]
```

# isin()

#### Arguments of isin() will be applied on each ROW of that COLUMN with '==' operator and return True for that ROW where it is satisfied

```python
condition = df_new["AreaName"].isin(["Asia", "Africa"])  #Argument need to provide as a List
```

```python
df_new[condition]
```

# Applying OR operator - Applying on Multiple Columns

```python
cond_1 = df_new["AreaName"] == "Asia"
cond_2 = df_new["AreaName"] == "Africa"
cond_3 = df_new["DevName"] == "Developing regions"
df_new[(cond_1 | cond_2) & cond_3]
```

# isin()

```python
df_new[(df_new["AreaName"].isin(["Asia", "Africa"])) & (df_new["DevName"].isin(["Developing regions"]))]
```
