```python
import pandas as pd
import xlrd
```

```python
df = pd.read_excel("Canada.xlsx")
df.columns.values
```

```python
df_1 = df[['AreaName', 1980]].head()
df_2 = df[['AreaName', 1980]].tail()
```

# Append Two Dataframe Containing Same number of Columns

#### Append will always append other DataFrame at the botton of Another

```python
df_1
```

```python
df_2
```

```python
df_1.append(df_2)  # df_2 will be appended at bottom of df_1. No change in Index Values
```

```python
df_1.append(df_2, ignore_index=True)  # Index will be re-arranged
```

# Append Two Dataframe Containing different number of Columns

#### All columns will be cosidered. If any column doesn't exist for a DataFrame, For that column values will be NULL

```python
df_3 = df[['AreaName', 1980, 1981]].head()
```

```python
df_2
```

```python
df_3
```

```python
df_2.append(df_3)
```

```python
df_2.append(df_3, ignore_index=True)
```

# ignore_index=True ---- Use Carefully

#### Ignore_index=True will make Index value 0 to n, regardless Index are String type. So while using this argument, use it carefully

```python
df_1.set_index("AreaName", inplace=True)
```

```python
df_2.set_index("AreaName", inplace=True)
```

```python
df_1
```

```python
df_2
```

```python
df_1.append(df_2)
```

```python
df_1.append(df_2, ignore_index=True)
```
